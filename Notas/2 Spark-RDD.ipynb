# RDD

La idea de Spark es dividir los datos grandes en partes ms pequeas y procesar cada una de estas partes en paralelo, aprovechando el nmero de ncleos (cores) disponibles para el procesamiento.

Si suponemos que tenemos 2 ncleos/procesadores de trabajo y queremos dividir un archivo de procesamiento de informacin en 4 partes, entonces **a estas 4 particiones se les conoce como RDD**; y se dividirn en partes iguales segn el nmero de ncleos, en este caso, cada ncleo procesara 2.

**RDD** Resilient Distributed Dataset. Estos son inmutables, lo que significa que si queremos realizar un cambio en estas particiones, entonces tendramos que empezar desde cero, creando/asignando nuevos ncleos y creando nuevas particiones. Estas son como las particiones dell archivo en cuestion

Los RDD son ``fault-tolerant``, lo que implica que si se pierde alguna de estas particiones, esta se recuperar a travs de **DAG**.

**DAG** significa Directed Acyclic Graph. Imagina que en cada paso que se realiza una nueva operacin con los RDD, se guarda una copia del proceso y si, por algn motivo, se pierde algo en el camino, se recupera a s mismo utilizando el DAG, mirando hacia el paso anterior.

# Operaciones de RDD

 * Transformation: Este proceso utiliza los RDD para realizar algun cambio con ellos de base y genera nuevos RDD. Los RDD *no son human readable*

 * Action: Este proceso utiliza los RDD para realizar una accin con ellos. Los RDD *no son human readable* pero con el comanto action los volvemos legibles, algo asi como un printout de los RDD para observarlo. Es como hacerlee a un display a los RDD
