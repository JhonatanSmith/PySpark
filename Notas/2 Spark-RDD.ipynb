# RDD

La idea de Spark es dividir los datos grandes en partes más pequeñas y procesar cada una de estas partes en paralelo, aprovechando el número de núcleos (cores) disponibles para el procesamiento.

Si suponemos que tenemos 2 núcleos/procesadores de trabajo y queremos dividir un archivo de procesamiento de información en 4 partes, entonces **a estas 4 particiones se les conoce como RDD**; y se dividirán en partes iguales según el número de núcleos, en este caso, cada núcleo procesaría 2.

**RDD** Resilient Distributed Dataset. Estos son inmutables, lo que significa que si queremos realizar un cambio en estas particiones, entonces tendríamos que empezar desde cero, creando/asignando nuevos núcleos y creando nuevas particiones. Estas son como las particiones dell archivo en cuestion

Los RDD son ``fault-tolerant``, lo que implica que si se pierde alguna de estas particiones, esta se recuperará a través de **DAG**.

**DAG** significa Directed Acyclic Graph. Imagina que en cada paso que se realiza una nueva operación con los RDD, se guarda una copia del proceso y si, por algún motivo, se pierde algo en el camino, se recupera a sí mismo utilizando el DAG, mirando hacia el paso anterior.

# Operaciones de RDD

 * Transformation: Este proceso utiliza los RDD para realizar algun cambio con ellos de base y genera nuevos RDD. Los RDD *no son human readable*

 * Action: Este proceso utiliza los RDD para realizar una acción con ellos. Los RDD *no son human readable* pero con el comanto action los volvemos legibles, algo asi como un printout de los RDD para observarlo. Es como hacerlee a un display a los RDD
